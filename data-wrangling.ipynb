```python
#Para poder manejar mejor los datos, asignamos valor de cero a los nulos y tambien a las celdas que no tienen  asignado un href
href_data[["href","href_formula"]]=href_data[["href","href_formula"]].fillna(0)
href_data["href"][365]=0
href_data["href_formula"][448]=0
href_data
```

```python
#Vamos a checar que links de la columna href dan error para despues intercambiarlos por el link en la columna href_formula

stat_href=[]
for link in href_data["href"]:
    if link!=0:
        html=requests.get(link)
        if html.status_code==404:
            stat_href.append("error")
        else:
            stat_href.append( "bien")
    else:
        stat_href.append("nan")
```

```python
#cree otra columna con el estado del link de href para visualizar mejor los datos a investigar
href_data["stat_href"]=stat_href
href_data
```

```python
#Asignamos el link de href_formula a las celdas de la columna href donde stat_href es error o nulo 
href_data.loc[(href_data["stat_href"]=="error")|(href_data["stat_href"]=="nan"),"href"]=href_data["href_formula"]
lista=href_data["href"].index
```

```python
#Cambiamos los valores de href necesarios en el data frame original con ayuda de una lista de indices donde 
#queremos hacer los cambios
data.loc[lista,"href"]=href_data["href"]
```
